Un bloque cuyo peso es W se desliza sobre un plano inclinado lubricado por una película de aceite como se indica en la Figura P1.45. La superficie de contacto del bloque es A y el espesor de la película de aceite *h*. Suponiendo una distribución lineal de velocidad en el aceite, halle una expresión para la velocidad "límite" *V* (aceleración nula) del bloque. Calcule la velocidad límite del bloque si la masa del mismo es 6kg, *A = 35 cm^2* , *theta = 15º* y la película lubricante es de aceite SAE 30 a 20ºC y tiene 1mm de espesor. 



### Datos
- Peso bloque = 6 kg
- Área de contacto = 35 cm^2
- Theta = 35º
- Espesor película aceite = 1 mm
- Aceite SAE 30
- Temperatura aceite = 20º
- Viscosidad aceite = 0,29 kg/(m*s)
 
### ¿Qué nos piden?
- Velocidad límite del bloque si sigue una distribución lineal

### Cálculo
###### Sumatorio de fuerzas en x 
- N+Wcos(theta)=0

###### Sumatorio de fuerzas en y
- A*tau -W*sin(theta)=0

###### Distribución de la velocidad lineal
Donde, en este caso podemos deducir que du=V y dy=h
- tau=viscosidad*(du/dy)

###### Esfuerzo cortante
- tau=Fcortante/A

```python
from math import *
A=0.0035 #m^2
P=6 #kg
W=P*9.81
theta=35*(2*pi/360) #radianes
viscosidad=0.29 #kg/(m*s)
h=0.001 #m
#  Equilibrio de fuerzas en y =0
#  Obtenemos que A*tau-W*sin(theta)=0
#  Como sabemos que tau=(du/dy)*viscosidad podemos obtener V
#  Aislamos la V y obtenemos la siguiente ecuación

V=(W*h*sin(theta))/(A*viscosidad)
print('La velocidad limite de nuestro cuerpo es V= %4.3g m/s' %V)
```

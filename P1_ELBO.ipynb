Un bloque cuyo peso es W se desliza sobre un plano inclinado lubricado por una pelcula de aceite como se indica en la Figura P1.45. La superficie de contacto del bloque es A y el espesor de la pelcula de aceite *h*. Suponiendo una distribucin lineal de velocidad en el aceite, halle una expresin para la velocidad "lmite" *V* (aceleracin nula) del bloque. Calcule la velocidad lmite del bloque si la masa del mismo es 6kg, *A = 35 cm^2* , *theta = 15* y la pelcula lubricante es de aceite SAE 30 a 20C y tiene 1mm de espesor. 



### Datos
- Peso bloque = 6 kg
- rea de contacto = 35 cm^2
- Theta = 35
- Espesor pelcula aceite = 1 mm
- Aceite SAE 30
- Temperatura aceite = 20
- Viscosidad aceite = 0,29 kg/(m*s)
 
### Qu nos piden?
- Velocidad lmite del bloque si sigue una distribucin lineal

### Clculo
###### Sumatorio de fuerzas en x 
- N+Wcos(theta)=0

###### Sumatorio de fuerzas en y
- A*tau -W*sin(theta)=0

###### Distribucin de la velocidad lineal
Donde, en este caso podemos deducir que du=V y dy=h
- tau=viscosidad*(du/dy)

###### Esfuerzo cortante
- tau=Fcortante/A

```python
from math import *
A=0.0035 #m^2
P=6 #kg
W=P*9.81
theta=35*(2*pi/360) #radianes
viscosidad=0.29 #kg/(m*s)
h=0.001 #m
#  Equilibrio de fuerzas en y =0
#  Obtenemos que A*tau-W*sin(theta)=0
#  Como sabemos que tau=(du/dy)*viscosidad podemos obtener V
#  Aislamos la V y obtenemos la siguiente ecuacin

V=(W*h*sin(theta))/(A*viscosidad)
print('La velocidad limite de nuestro cuerpo es V= %4.3g m/s' %V)
```

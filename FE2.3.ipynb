## PROBLEMA FE2.3
En la figura FE2.2, si el aceite de la región B tiene una densidad relativa ρT = 0.8 y la presión absoluta en el punto A es 1 atm, ¿cuál es la presión absoluta en el punto B? 



En este ejercicio nos piden la presión en el punto B.     
Vamos a buscar la relación entre la Pa y Pb.    
Por lo tanto si empezamos en el punto A sabemos que:

###### Pa + SGw * γw * 0,05 + SGm * γm * (0,08-0,04) - SGoil * γw * 0,03 = Pb

Donde γw = 9790 N/m, SGw = 1, SGm = 13.56 y SGoil = 0.8. 


```python
#Agua
γw = 9790 #N/m
SGw = 1
Zw = 0.05 #m

#Mercurio
SGm = 13.56
Zm = 0.08 - 0.04 #m

#Aceite
SGoil = 0.8
Zoil = 0.03
Pa = 101300 #Pascals

Pb = Pa + SGw*γw*Zw + SGm*γw*Zm - SGoil*γw*Zoil
print('La presión absoluta en el punto B es %4.3g Pascals' %Pb)
```

## PROBLEMA PE2.7
Venecia, en Italia, se hunde lentamente, por eso, sus plazas y calles se inundan durante las tormentas, especialmente en invierno. La solución propuesta es la del dique flotante de la Figura PE2.7. Cuando se llena de aire, se eleva por encima del nivel del mar. El dique tiene 30m de alto, 5m de ancho y 20m de profundidad. Suponga una densidad uniforme de 300kg/m^3 cuando flota. Para la diferencia entre el mar y la laguna de 1m, estime el ángulo al cual flota. 


**Fuerzas Aplicadas**     
Primero de todo debemos tener en cuenta que fuerzas estan actuando en nuestro cuerpo.   
En este caso hay cuatro:
- La fuerza hidroestática producida por el Mar Adriático
- La fuerza hidroestática producida por la Laguna de Venecia
- El peso del dique
- La fuerza sustentadora de la parte sumergida en el agua

1. Debemos encontrar las partes sumergidas en ambos lados del dique con el objetivo de encontrar las dos fuerzas hidroestaticas
    - Si θ es el ángulo que forma el eje y dirección positiva con el dique podemos deducir que: 
    - distancia_Mar_Adriático = 25 m / cos(θ)
    - distancia_Laguna = 24 m / cos(θ)
    
2. Necesitamos encontrar las fuerzas que se estan aplicando en el dique
    - F_Mar_Adriático = ρ_water * g * distancia_Mar_Adriático * altura_MA
    - F_Laguna = ρ_water * g * distancia_Laguna * altura_L
    - W = ρ_dique * g * Longitud * anchura_dique
    - B = ρ_water * g * Longitud_dique_sumergido * anchura_dique
    
3. Finalmente si calculamos los momentos respecto a la articulación de la pieza, deben ser igual a cero. Por lo tanto podemos calcular el resultado final.
    - Momento_CG = 0
    - (F_Mar_Adriático * dist_CG1 - F_Laguna * dist_CG2 )cos(θ) + (W * dist_CG2  - B * dist_CG4)sin(θ) = 0

Aislamos y obtenemos el ángulo   
##### θ = math.asin((- F_Mar_Adriático * dist_CG1 + F_Laguna * dist_CG2) / (W * dist_CG3 - B * dist_CG4))

```python
import math
ρ_water = 1025 #Kg/m^3
ρ_dique = 300 #Kg/m^3

Longitud_dique = 30 #m
altura_L = 24/2 #m
altura_MA = 25/2 #m
anchura_dique = 5 #m 
Longitud_dique_sumergido = (24+25)/2

g = 9.81 #m/s
distancia_Mar_Adriático = 25 # /cos(θ)
distancia_Laguna = 24 # / cos(θ)

F_Mar_Adriático = ρ_water * g * altura_MA * distancia_Mar_Adriático # /cos(θ)
F_Laguna = ρ_water * g * altura_L * distancia_Laguna # /cos(θ)
W = ρ_dique * g * Longitud_dique * anchura_dique
B = ρ_water * g * Longitud_dique_sumergido * anchura_dique

dist_CG1 = 25/3 #*cos(θ)
dist_CG2 = 24/3 #*cos(θ)
dist_CG3 = 15 #*sin(θ)
dist_CG4 = anchura_dique_sumergido/2 #*sin(θ)

θ = (360/(math.pi*2))*math.asin((-F_Mar_Adriático*dist_CG1 + F_Laguna*dist_CG2)/(W*dist_CG3 -B*dist_CG4))
print('El ángulo necesario para que flote es %4.3g grados' %θ)
```

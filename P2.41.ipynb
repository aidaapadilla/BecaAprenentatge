## PROBLEMA P2.41
El sistema de la figura P2.41 esta a 20ºC.      
Calcule la presión absoluta en el punto A en lbf/ft^2


Primero de todo tenemos que mirar en las tablas del white para encontrar los pesos especificos del agua y del mercurio. 


1. Como nos piden el resultado en lbf/ft^2 vamos a cojer el peso especifico en esa unidad:    
  - γw = 62.4 lbf/ft^3
  - γm = 846 lbf/ft^3  
  
2. Posteriormente aplicamos la formula de la hidroestatica desde el punto A hasta la superficie
  - Es importante tener en cuenta que siempre que nos dan SG, nos estan proporcionando la relación de ese fluido con el peso específico del agua, por lo tanto debemos multiplicarlo por ese valor. 
  - Consideramos z positivo en eje y negativo. 

#### Pressure_A + SGoil * γw * Zoil - γm * Zm + γw * Zw = pa

Es importante tener en cuenta que pa debe estar en lbf/ft^2 y que las distancias también deben estar en ft. 
Para pasar de inches a ft dividimos entre 12. 


```python
#Aceite
SGoil = 0.85
Zoil = 6/12 #ft

#Mercurio
γm = 846 #lbf/ft^3 
Zm = 10/12 #ft

#Agua
γw = 62.4 #lbf/ft^3
Zw = 5/12 #ft

#Presión atmosferica
pa = 14.7 * (12*12) #lbf/ft^2

#Aislamos la Pressure_A de la ecuación de arriba y obtenemos: 
Pressure_A = pa -SGoil*γw*Zoil + γm*Zm - γw*Zw 
print('La presión en el punto A es %4.3g lbf/ft^2' %Pressure_A)
```

## PROBLEMA 2.87

La botella de champan (SG=0.87) de la figura P2.87 está presurizada según indica la lectura del manómetro de mercurio. Calcule al fuerza neta sobre la base semiesférica de la botella si tiene un radio de 2 inches. 



1. Primero de todo vamos a calcular la presión a la que esta sometida en la parte inferior de la botella aplicando la fórmula de la hidroestática
  - Para hacerlo vamos a necesitar el peso especifico del agua, ya que nos proporcionan el SG del champán.
  - γw = 62.4 lbf/ft^3

#### P_bottom + SGchampan * γw * Zchampan - SGmercurio * γw * Zmercurio = Patm

2. Una vez tenemos el valor de la presión en el fondo de la botella vamos a calcular la fuerza aplicada. 
  - Vamos a tener en cuenta únicamente las fuerzas verticales por lo tanto vamos a necesitar: 
    - Area_bottom = π/4 * r^2
    - Peso_cilindro = SGchampan * γw * π * r^2 * h
    - Peso_semiesfera = SGchampan * γw * 2π/3 * r^3
    
    
3. Finalmente sabemos que la fuerza aplicada será:

#### Fv = P_bottom * Area_bottom + Peso_cilindro - Peso_semiesfera

```python
import math
# DATOS
SGchampan=0.96
SGmercurio=13.56  #Lo obtenemos de la tabla 2.1 del white
γw = 62.4 #lbf/ft^3
Zchampan = 2/12 #ft
Zmercurio = 4/12 #ft
r=2 /12 #ft
Patm = 0

#Calculamos la presión en el fondo de la botella
P_bottom = Patm-SGchampan*γw*Zchampan + SGmercurio*γw*Zmercurio

#Ahora calculamos la fuerza vertical
h=6/12 #ft
r_bottom=4/12 #ft

Area_bottom = (math.pi/4)*r_bottom**2
Peso_cilindro = SGchampan*γw*math.pi*(r**2)*h
Peso_semiesfera = SGchampan*γw*(2*math.pi/3)*(r**3)

Fuerza_vertical = P_bottom*Area_bottom + Peso_cilindro - Peso_semiesfera

print('La fuerza vertical aplicada es de %4.3g lbf/ft^2' %Fuerza_vertical)
```

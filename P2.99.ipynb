## PROBLEMA 2.99 
Una esfera de 2ft de diámetro pesa 400 lbf y cierra un orificio de 1 ft de diámetro en l aparte inferior del depósito de la Figura P2.99. Calcule la fuerza F requerida para desplazar la esfera del agujero. 



En este problema nos piden que calculemos la fuerza F dibujada.     
Por lo tanto, nos han dado el peso de la esfera. 
- W = 400 lbf   
Y sabemos que el volumen de una esfera es:
- V = 4π/3 * R^3    
Por lo tanto si calculamos el sumatorio de fuerzas en el eje *Y* podemos encontrar la fuerza que seria necesaria para desplazar la esfera. 
1. Fuerzas actuando hacia arriba
    - F_upwards = γ_water * Volumen
2. Fuerzas acuando hacia abajo
    - F_downwards = γ_water * A_forat * altura
    - A_forat = π * (R/2)^2
    - altura = R*3
3. Fuerza necesaria para mover la esfera: 
    - F = W + F_downwards - F_upwards

```python
import math
W = 400 #lbf
R = 1 #ft
γ_water = 62.4 
Volumen = (4*math.pi/3) * R**3
Volum_forat = (4*math.pi/3) * (0.688)**3
A_forat = math.pi * (R/2)**2
altura = R*3

F_upwards = γ_water * Volumen
print(F_upwards)
F_downwards =   γ_water * A_forat * altura
print(F_downwards)

F_mover_esfera = W + F_downwards - F_upwards
print('La fuerza necesaria para mover la esfera sera de %4.3g lbf' %F_mover_esfera)
```

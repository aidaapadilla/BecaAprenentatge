## PROBLEMA 1.56
El dispositivo de la Figura P1.56 se denomina *cosímetro cono-placa*. El ángulo del cono es muy pequeño, por lo que el sin(θ)=θ, y el juego entre cono y placa se llena con el liquido a ensayar, midiendo el par M que hay que aplicar para hacer girar el cono a la velocidad Ω. Suponiendo un perfil de velocidad lineal en la película fluida, obtenga una expresion para la viscosidad del fluido μ en función de (M, R, Ω, θ). 


Primero de todo vamos a buscar el trozo de liquido que queremos extraer
- Sabemos que tan(θ) = h/r, por lo tanto podemos deducir que h=r*tan(θ)

Una vez tenemos esta relación, vamos a calcular el torque de la figura donde: 
- d(Torque) = dM = τ*dA*r
- τ = μ*Ω*r/r*tan(θ)
- dA = 2π*r*(dr/cos(θ)) 

Integramos en los dos en ambos lados y obtenemos: 
- M = [2πμΩ/(tan(θ)*cos(θ))] * ∫r^2 *dr
- Los límites de la integral van de 0 a R. 

Sabemos que tan(θ) = sin(θ)/cos(θ), por lo tanto M valdrá:
- M = [2πμΩ/sin(θ)] * R^3/3

Obtenemos la expresión de la viscosidad aislando de la ecuación μ (Aplicando un angulo pequeño *sin(θ)= θ* )
##### μ = 3Mθ / (2πΩR^3)

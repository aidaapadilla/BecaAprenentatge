# Problema 1.53
Un cono sólido de ángulo *2θ*, radio de la base *Ro* y densidad *ρc*, está girando con una velocidad angular *ωo* en su asiento cónico, como se muestra en la Figura P1.53.
La holgura h entre ambos se llena con aceite de viscosidad *μ*. Despreciando la resistencia del aire, **derive una expresión analítica para la velocidad angular del cono *ω(t)*** si no hay par aplicado.



La ecuación que nos relaciona el torque con la velocidad angular es la siguiente: 
#### M = -I0 * dw/dt

Por lo tanto debemos calcular el torque y el momento de inercia de nuestro cuerpo con el objetivo de obtener w(t)

1. Primero de todo vamos a calcular el par aplicado
    ##### - d(Torque) = dM = τ dA r
    - τ = μwr/h
    - dA = 2πr*(dr/cos(θ)) 

    Finalmente obtenemos: 
    - ∫dM = 2πμw/(hsin(θ))* ∫r^3 *dr

    Donde los límites de integración de dM son (0,M) y de dr (0, r0)
    Obtenemos que el torque será:
    ##### M= πμwR^4 / 2hsin(θ)

2. Ahora procedemos a calcular I0
    Buscando en una tabla de momentos de inercia encontramos el de un cono: 
    ##### I0 = 3/10 * m r0^2 
    - Donde *m* es la masa del cuerpo y *r0* el radio del cono

3. Substituimos y calculamos w(t)
    πμwR^4 / 2hsin(θ) = - 3/10 * m r0^2 *(dw/dt)
    - Separamos las variables que dependen de w y las que dependen del tiempo y integramos
    ∫ dw/w = - πμwR^2 / 2hmhsin(θ) *∫ dt
    - Donde los limites de integración de *dw* son (w0, w) y de *dt* (0,t)
    ln(w/w0) = - πμwR^2 * t / 2hmhsin(θ)
    
Finalmente el resultado es: 
### w = w0 * exp[- πμwR^2 * t / 2hmhsin(θ)]

# Problema 1.53
Un cono slido de ngulo *2*, radio de la base *Ro* y densidad *c*, est girando con una velocidad angular *o* en su asiento cnico, como se muestra en la Figura P1.53.
La holgura h entre ambos se llena con aceite de viscosidad **. Despreciando la resistencia del aire, **derive una expresin analtica para la velocidad angular del cono *(t)*** si no hay par aplicado.



La ecuacin que nos relaciona el torque con la velocidad angular es la siguiente: 
#### M = -I0 * dw/dt

Por lo tanto debemos calcular el torque y el momento de inercia de nuestro cuerpo con el objetivo de obtener w(t)

1. Primero de todo vamos a calcular el par aplicado
    ##### - d(Torque) = dM =  dA r
    -  = wr/h
    - dA = 2r*(dr/cos()) 

    Finalmente obtenemos: 
    - dM = 2w/(hsin())* r^3 *dr

    Donde los lmites de integracin de dM son (0,M) y de dr (0, r0)
    Obtenemos que el torque ser:
    ##### M= wR^4 / 2hsin()

2. Ahora procedemos a calcular I0
    Buscando en una tabla de momentos de inercia encontramos el de un cono: 
    ##### I0 = 3/10 * m r0^2 
    - Donde *m* es la masa del cuerpo y *r0* el radio del cono

3. Substituimos y calculamos w(t)
    wR^4 / 2hsin() = - 3/10 * m r0^2 *(dw/dt)
    - Separamos las variables que dependen de w y las que dependen del tiempo y integramos
     dw/w = - wR^2 / 2hmhsin() * dt
    - Donde los limites de integracin de *dw* son (w0, w) y de *dt* (0,t)
    ln(w/w0) = - wR^2 * t / 2hmhsin()
    
Finalmente el resultado es: 
### w = w0 * exp[- wR^2 * t / 2hmhsin()]
